```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, accuracy_score
```

```python
def convert_ent_type(ent_type):
    if ent_type == 'I-ORG':
        return 'ORG'
    elif ent_type == 'I-LOC':
        return 'LOC'
    elif ent_type == 'I-PER':
        return 'PER'
    elif ent_type == 'I-MISC':
        return 'MISC'
    elif ent_type == '0':
        return ''
```

```python
df = pd.read_csv('hun_ner_corpus.txt', names=['token', 'ent_type'], skiprows=1, sep='\t', encoding='latin-1') # https://rgai.inf.u-szeged.hu/file/67#overlay-context=
sample = df.iloc()[:10000].copy()
ent_types = list(set(sample['ent_type'])))
sample['ent_type'] = sample.apply(lambda x: ent_types.index(x['ent_type'])), axis=1)
chunk_size = 50
chunk_overlap = 10

data = []
for idx in range(0, 10000, chunk_size - chunk_overlap):
    chunk = ' '.join(list(sample[idx:idx+chunk_size]['token'].values))
    ent_types = list(sample[idx:idx+chunk_size]['ent_type'].values)
    data.append((chunk, ent_types))

dataset = pd.DataFrame(data, columns=['chunk', 'ent_types'])

dataset['preds'] = dataset.apply(lambda x: list(map(lambda y: y.ent_type_, nlp(x['chunk']))), axis=1)
dataset.head()
```

```python
y = list(dataset['ent_types'].values)
preds = list(dataset['preds'].values)
print(len(y),len(preds))

report = classification_report(y, preds, zero_division=np.nan)
print(report)
```
